```python
import pandas as pd
import plotly.express as px
```

```python
from google.colab import files
import io
uploaded=files.upload()
```

```python
df_titanic=pd.read_csv(io.BytesIO(uploaded['train.csv']))
df_titanic
```

```python
#1
print("il y'avait",(df_titanic['Age']>40).sum(),"personnes de plus de 40 ans au bord du titanic")
```

```python
#2
print("il y'avait",round((((df_titanic['Survived']==1).sum())*100)/len(df_titanic['Survived']),1),"% de survivants")
```

```python
#3
def en_num(Sex):
  if Sex=='male':
    return 0
  if Sex=='female':
    return 1

df_titanic['Sex_class']=df_titanic['Sex'].apply(lambda Sex : en_num(Sex))
df_titanic
```

```python
#4
def en_caractere(Pclass):
  if Pclass==3:
    return 'troisieme classe'
  if Pclass==2:
    return 'deuxieme classe'
  if Pclass==1:
    return 'première classe'

df_titanic['Pclass_string']=df_titanic['Pclass'].apply(lambda Pclass:en_caractere(Pclass))
df_titanic
```

#5
On en déduit que les tarifs sont plus élevés en première classe, et que les tarifs en classe deuxième sont légerement plus élévés que ceux en troisième classe.

```python
#5
fig=px.histogram(df_titanic,x='Fare',y='Pclass_string',histfunc='avg',title='Histogramme de la classe du passager en fonction du tarif moyen')
fig.show()
```

#6
On en déduit que le taux de survivants est approximativement le même pour chaque classe, il n'y a pas un énorme écart : 136 en première classe,87 en deuxième et 119 en troisème.
Cependant, puisqu'il y'avait plus de passagers en troisième classe, il y'avait plus de non surivants.
Le fait qu'il y'ai moins de survivants qu'en première classe (alors qu'ils sont beaucoup moins nombreux) peut indiquer que les passagers en première étaient favorisés.

```python
#6
fig=px.histogram(df_titanic,'Pclass_string',color='Survived',title='histogramme de la classe des passagers colorisé par si ils ont survit ou non')
fig.show()
```

#7
Il y'avait plus de passagers hommes que femmes, mais plus de survivants femmes que hommes, cela indique que les femmes étaient favorisés.

```python
#7
fig=px.histogram(df_titanic,'Sex',color='Survived', title='histogramme du sexe des passagers colorisé par si ils ont survit ou non')
fig.show()
```

#8
Pour faire un algorithme de prédiction de survivabilité au naufrage du titanic, je séléctionnaires le sexe, la classe des passagers et le tarif puisqu'il est indicateur de la classe du passager.

```python
#9
from sklearn.model_selection import train_test_split
```

```python
x=df_titanic[['Sex_class','Pclass','Fare']]
y=df_titanic['Survived']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=250)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier()
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
```
