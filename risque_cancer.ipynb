```python
#question
#ajouter une colonne 'risklevel_class' qui represente le risk par chiffre 0, 1 ou 2
#on crée une fonction pour changer le risque en numéro
def change_num(categorie):
  if categorie == 'low risk':
    return 0
  elif categorie == 'mid risk':
    return 1
  else:
    return 2
```

```python
#on doit donc lui appliquer la fonction change_num
#on utilise donc la fonction '.apply(lambda categorie[ou x] : le nom de la fonction())
df_maternal['Risklevel_class'] = df_maternal['RiskLevel'].apply(lambda categorie : change_num(categorie))
df_maternal
```

```python
#question
#le but encore une fois c'est ajouté une nouvelle colonne 'niveau risque string'
#cette colonne va représenter le risque par une chaine de caractere en français. en gros on veut que les modalités 0,1,2 deviennent des mots en francais
#on veut donc remplacer le mid risk qui vaut 1 par exemple par risque moyen
#on crée donc une nouvelle fonction
def change_classe(num):
  if num == 0:
    return 'risque faible'
  elif num == 1:
    return 'risque moyen'
  else:
    return 'risque eleve'
```

```python
#on l'applique
df_maternal['niveau_risque_string'] = df_maternal['Risklevel_class'].apply(lambda num : change_classe(num))
df_maternal
```

```python
from sklearn.model_selection import train_test_split
```

```python
#question 8
#a partir du tableau de données, on veut séparer un jeu de données d'entraînement de 90% et de test de 10 %
#selon l'énoncé, la valeur a prédire c'est le risque donc c'est le y
y = df_maternal['RiskLevel']  # entraine à prédire y

#on oublie pas de mettre deux crochets sinon sa ne marche pas
x = df_maternal[['Age','pression_artérielle_diastolique','pression_artérielle_systolique', 'BS', 'temperature_corporelle', 'rythme_cardiaque']]  #tester x pour prédire y
#on précise quel partie est test et train
#x avant y et train avant test
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.1)

#pour vérifier
x_train
```

```python
#question 9
#on importe une autre fonction
from sklearn.naive_bayes import GaussianNB
```

```python
#on veut entrainer le modèle
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
#on veut tester la prédiction
#on veut évaluer la pérforlance du modéle
#ne pas se tromper pour predire on prend x_test
prediction = model.predict(x_test)
```

```python
#pour preciser le nbre de prédiction a faire on utilise [:5]
prediction [:5]
print(prediction [:5])

#on va lui dire de print le truc en lui meme
print(y_test[:5])
```

```python
#methode
#on importe de nouvelles fonctions
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
#nous donne combien le modéle est fiable
#ici on utilise y_test parceque on le fait sur les vraies valeurs
accuracy_score(prediction, y_test)
print('ce kk est fiable à ;', accuracy_score(prediction, y_test))
```

```python
#la classification report calcule la précision pour chaque modalité
classification_report(prediction, y_test)
print(classification_report(prediction, y_test))
```
